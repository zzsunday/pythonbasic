```python
import pickle
```

#dump:不落地
#dumps：落地

```python
lst='a b c'.split()
```

```python
lst
```

```python
d=dict(zip('abc',range(3)))
```

```python
with open('bin','wb+') as f:
    pickle.dump(lst,f)
    pickle.dump(d,f)
```

```python
with open('bin','rb+') as f:
    tmp=pickle.load(f)
    print(type(tmp))
    print(tmp)
```

```python
class AA:
    def show(self):
        print('abc')
```

```python
aa=AA()
```

```python
with open('bin','wb+') as f:
    pickle.dump(aa,f)
```

```python
with open('bin','rb+') as f:
    tmp=pickle.load(f) #存进去的是AA
    print(type(tmp))
```

```python
tt=tmp#重新创建一个实例 
```

```python
tt.show()
```
