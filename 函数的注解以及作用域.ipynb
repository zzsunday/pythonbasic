## 函数本地作用域

```python
def foo(xyz=[],m=5,n=6):
    xyz.append(100)
    print(xyz)
```

```python
print(1,foo.__defaults__)
print(foo(),id(foo))
print(2,foo.__defaults__)
print(foo(),id(foo))
print(3,foo.__defaults__)
```

```python
def foo(xyz=[],u='abc',z=123):
    xyz=xyz[:]  #做了一个拷贝，就相当于新建一个xyz
    xyz.append(1)
    print(xyz)
foo()
print(1,foo.__defaults__)
foo()
print(2,foo.__defaults__)
foo([10])
print(3,foo.__defaults__,sep=':')
foo([10,5])
print(4,foo.__defaults__,sep=':')
```

```python
def foo(xyz=None,u='abc',z=123):
    if xyz is None:
        xyz=[]
    xyz.append(1)
    print(xyz)
foo()
print(1,foo.__defaults__)
foo()
print(2,foo.__defaults__)
foo([10])
print(3,foo.__defaults__)
foo([10,5])
print(4,foo.__defaults__,sep=':')
lst=[20]
foo(lst)
```

```python
def foo(xyz=None,u='abc',z=123):
    if xyz is None:
        xyz=[]
    xyz.append(1)
    return(xyz)
lst=foo()[1]
a=foo(lst)
print(a)
```

```python
def foo(xyz=[],u='abc',z=123):
    xyz.append(1)
    return xyz
def foo(xyz=[],u='abc',z=1234):
    xyz.append(1)
    return xyz
print(foo(),id(foo),foo.__defaults__)

del foo

print(foo(),id(foo),foo.__defaults__)
```

```python
def foo(xyz=[],u='abc',z=123):
    xyz.append(1)
    def inner(a=10):
        pass
    print(inner)
    def inner(a=100):
        print(xyz)
    print(inner)
    return inner

bar=foo()
print(id(foo),id(bar),foo.__defaults__,bar.__defaults__)
```

```python
del bar
print(id(foo),id(bar),foo.__defaults__,bar.__defaults__)
```
