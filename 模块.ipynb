##    __slots__

```python
class A:
    X = 123
    __slots__ = ('p1',)
    def __init__(self):
        self.p1 = 1
        self.p2 = 2
    def showme(self):
        print('I AM A.{},{}'.format(self.p1, self.p2))
```

```python
print(A.__dict__)
print(A().__dict__)
```
