### defalutdict

```python
from collections import defaultdict
import random
```

```python
d1={}
d2=defaultdict(list) #初始化
for k in 'abcde':
    for v in range(5):
        if k not in d1.keys():
            d1[k]=[]
        d1[k].append(v)
print(d1)
```

```python
for k in 'mnopq':
    for v in range(3):
        d2[k].append(v)
print(d2)
```

```python
d1 = {}
for k in 'abcdef':
    for i in range(random.randint(1,5)):
        if k not in d1.keys():
            d1[k] = []
        d1[k].append(i)
print(d1)
d1 = defaultdict(list)
for k in 'abcdef':
    for i in range(random.randint(1,5)):
        d1[k].append(i)
print(d1)
```

## ordereddict

```python
from collections import OrderedDict
import random
d = {'banana': 3, 'apple': 4, 'pear': 1, 'orange': 2}
print(d)
keys = list(d.keys())
random.shuffle(keys)
print(keys)
od = OrderedDict()
for key in keys:
    od[key] = d[key]
print(od)
print(od.keys())
```

### 用户输入一个数字，打印每一位数字及其重复的次数

```python
while True:
    num=input(">>>").strip().lstrip('0')
    if num.isdigit():
        break
    else:
        print('input again')
m=len(num)
d={}
for c in num:
    if not d.get(c):
        d[c]=1
    else:
        d[c]+=1
```

```python
while True:
    num=input(">>>").strip().lstrip('0')
    if num.isdigit():
        break
    else:
        print('input again')
m=len(num)
d={}
for c in num:
    if not d.get(c):
        d[c]=0
    d[c]+=1
```
