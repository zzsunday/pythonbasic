```python
import threading
import time
def worker1():
    for _ in range(10):
        time.sleep(0.5)
        print('1....................')
    print('1==================')
    
def worker2():
    count = 0
    for _ in range(10):
        count += 1
        time.sleep(0.5)
        print('2....................')
        if count > 5:
            break
    print('2==================')

    
print(threading.current_thread())
t = threading.Thread(target=worker1)
t.start()

t = threading.Thread(target=worker2)
t.start()
```

```python
import threading
import time
def worker1():
    print('inside',threading.current_thread())
    print('inside',threading.main_thread())
    print('inside',threading.active_count())
    for _ in range(5):
        time.sleep(0.5)
        print('1....................')
    print('1==================')

print('outer',threading.current_thread())
t = threading.Thread(target=worker1)
t.start()
```
