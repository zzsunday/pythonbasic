```python
lst1=[1,1]
n=2
while True:
    lst=[]
    for i in range(n-1):
        num=lst1[i]+lst1[i+1]
        lst.append(num)

    lst.append(1)
    lst.insert(0,1)
    for j in lst:
        print(j,end=' ')
    print()
    if n==10:
        break
    lst1=lst
    n+=1
        
```

```python
triangle=[[1], [1,1]]
for i in range(2,6):
    cur = [1]
    pre = triangle[i-1]
    for j in range(len(pre)-1):
        cur.append(pre[j]+pre[j+1])
    
    cur.append(1)
    triangle.append(cur)
print(triangle)
```

```python
triangle=[]
n = 6
for i in range(n):
    row = [1]
    triangle.append(row)
    if i==0:
        continue
    for j in range(i-1):
        row.append(triangle[i-1][j]+triangle[i-1][j+1])
    row.append(1)
print(triangle)
```

```python
import datetime
n=100000
pn=[]
count=0
start=datetime.datetime.now()
for i in range(2,n):
    for j in pn:
        count+=1
        if i%j==0:
            break
    else:
        pn.append(i)
delta=(datetime.datetime.now()-start).total_seconds()
print(delta)
# print(pn)
# print(count) #注意理解for else子句
        
```

```python
import math
primenumber = []
flag = False
count=0
start=datetime.datetime.now()
for x in range(2,100000):
    for i in primenumber:
        count+=1
        if x % i == 0 :
            flag = True
            break
        if i >= math.ceil(math.sqrt(x)):
            flag = False
            break
    if not flag:
        primenumber.append(x)
delta=(datetime.datetime.now()-start).total_seconds()
print(delta)
```

```python
lst=[]
n=6
for i in range(n):#n=0的时候不进入直接到1
    row=[1]
    for k in range(i):
        row.append(1) if k==i-1 else row.append(0)
    lst.append(row)
    if i==0:
        continue
    for j in range(1,i//2+1):
        val=lst[i-1][j-1]+lst[i-1][j]
        row[j]=val
        if j!=i-j:
            row[-j-1]=val
print(lst)
```
