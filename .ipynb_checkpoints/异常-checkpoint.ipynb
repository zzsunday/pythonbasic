```python
class MyException(Exception):
    pass

try:
    raise MyException()

except MyException:
    print('my exception')
except Exception:
    print('Exception')
except InterruptedError:
    print('InterruptedError')
```

```python
class MyException(Exception):
    pass

try:
    #1/0
    #open('tesa')
    raise MyException('My Error')

except MyException as e:
    print(e)
except Exception:
    print('Exception')
except InterruptedError:
    print('InterruptedError')
```

```python
try:
    1/0
    #open('tesa')
    raise MyException('My Error')

except MyException as e: #产生的异常和表示符对应起来
    print(e)
except Exception as e:
    print(e)
    print('Exception')
except InterruptedError:
    print('InterruptedError')
```

```python
class MyException(Exception):
    pass

try:
    #1/0
    #open('tesa')
    raise MyException

except MyException as e:
    print(e,type(e))
except Exception:
    print('Exception')
except InterruptedError:
    print('InterruptedError')
```

```python
class MyException(Exception):
    def __init__(self, xdata='mistake'):
        self.xdata = xdata

try:
    #1/0
    #open('tesa')
    raise MyException

except MyException as e:
    print('******')
    print(e,type(e))
except Exception as e:
    print('---')
    print(e,type(e))
except InterruptedError:
    print('InterruptedError')
```
