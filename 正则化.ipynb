```python
str1='''bottle\nbag\nbig\nable'''
```

```python
reg=re.compile('b(\w+)(e)')
res=reg.finditer(str1)
```

```python
for matcher in res:
    print(matcher.groups())
    print(matcher.group(0))
    print('----------------')
```

```python
reg=re.compile('b(?P<body>\w+)(?P<tail>e)')
res=reg.finditer(str1)
for matcher in res:
    print(matcher.groups())
    print(matcher.group(0))
    print(matcher.groupdict())
    print('----------------')
```
