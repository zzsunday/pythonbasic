## set

* 集合中不能存在不可变对象即不能有不能hash的数据类型  
* 集合效率很高，因为其遍历方式是遍历hash值

### 集合运算

#### 并集

```python
a={1,2,3}
b={4,5,6}
```

```python
c=a.union(b)
c
```

```python
a|b
```

```python
d=a.update(b)
print(d,a,sep='\n')
```

```python
a|={4,5}|{7,8}
a
```

#### 交集

```python
a={1,2,3}
b={3,4,5}
```

```python
a.intersection(b)
```

```python
a&b
```

#### 差集

```python
a={1,3,4,5,6}
b={3}
```

```python
a-b
```

```python
a-=b
a
```
