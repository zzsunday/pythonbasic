```python
import datetime
import time

def logger(t):# def logger(t1, t2, t3....tn):
    def _logger(fn):
        #@copy_properties(fn) 可以把上面写的复制属性的函数装饰在此
        def wrap(*args, **kwargs):
            #before 功能增强
            # print("args={},kwargs={}".format(args, kwargs))
            start = datetime.datetime.now()
            ret = fn(*args, **kwargs)
            #after 功能增强
            duration = (datetime.datetime.now() - start).total_seconds()
            if duration > t:
                print("function {} took {}s.".format(fn.__name__, duration))
            return ret
        return wrap
    return _logger

@logger(3)# add = logger(3)(add), @logger(3, 5, 9,...n)
def add(x, y):
    print("======call add======")
    time.sleep(5)
    return x + y

print(add(4, y=5))
```

### 写一个命令分发器
#### 程序员可以方便注册到某个命令，用户输入命令时注册该命令的函数
#### 如果此命令没有对应的注册函数，执行默认函数，用户输入用input(">")

```python
def cmds_dispatch():
    command={}
    def reg(name):
        def _reg(fn):
            command[name]=fn
            return fn
        return _reg
    def defaultfunc():
        print('Unknown command')

    def dispatch():
        while True:
            cmd=input('>>')
            if cmd.strip()== 'quit':
                return
            command.get(cmd,defaultfunc)()
    return reg,dispatch
reg,dispatch=cmds_dispatch() #本题精髓

@reg('corgi')
def foo1():
    print('welcome to the corgi da house')
@reg('py')
def foo2():
    print('python')

dispatch()
```
