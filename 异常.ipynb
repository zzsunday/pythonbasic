```python
def foo1():
    try:
        1/0
    finally:
        print('fin')

def foo2():
    try:
        foo1()
    finally:
        print('foo2 fin')

try:
    foo2()
except Exception as e:
    print(e)
    print(e.__class__)
print('outer')
```

```python
import time,threading
def foo1():
    try:
        1/0
    finally:
        print('foo1 fin')

def foo2():
    time.sleep(3)
    try:
        foo1()
    finally:
        print('foo2 fin')
        open('abcd')

t = threading.Thread(target = foo2)
t.start()

while True:
    time.sleep(1)
    print('================')
    if t.is_alive():
        print('alive')
    else:
        print('dead')
```

```python
def foo1():
    try:
        1/0
    finally:
        print('foo1 fin')

def foo2():
    time.sleep(3)
    foo1()
    
    while True:
        print('-------')
        time.sleep(1)
try:
    foo2()
except Exception as e:
    print(e)
    print(e.__class__)
print('outer')
```
